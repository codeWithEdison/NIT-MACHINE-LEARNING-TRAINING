```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter Plot
plt.figure()
plt.scatter(data['age'], data['genre'])
plt.xlabel('Age')
plt.ylabel('Genre')
plt.title('Age vs Genre')

# Line Plot
plt.figure()
plt.plot(data['age'])
plt.xlabel('Index')
plt.ylabel('Age')
plt.title('Age Distribution')

# Bar Chart
plt.figure()
data['genre'].value_counts().plot(kind='bar')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Genre Distribution')

# Histogram (you already have this)
plt.figure()
plt.hist(data['age'], color='skyblue', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Age')

# Box Plot
plt.figure()
sns.boxplot(x='genre', y='age', data=data)
plt.xlabel('Genre')
plt.ylabel('Age')
plt.title('Age Distribution by Genre')

# Heat Map
plt.figure()
correlation = data.corr()
sns.heatmap(correlation, annot=True)
plt.title('Correlation Heatmap')

plt.show()
```
